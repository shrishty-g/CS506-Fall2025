```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import distance
from sklearn.metrics import jaccard_score
from sklearn.metrics.pairwise import cosine_similarity
```

```python
x = np.array([1, 2, 3])
y = np.array([4, 6, 8])

print("x:", x)
print("y:", y)
```

```python
for p in [1, 2, 3]:
    d = distance.minkowski(x, y, p)
    print(f"Minkowski (p={p}): {d:.3f}")
```

```python

d1 = distance.cityblock(x, y)
print(f"Manhattan Distance: {d1:.3f}")

d2 = distance.euclidean(x, y)
print(f"Euclidean Distance: {d2:.3f}")


dinf = distance.chebyshev(x, y)
print(f"Chebyshev Distance: {dinf:.3f}")
```

```python
#Jaccard Similarity next


A = np.array([1, 1, 0, 1])
B = np.array([1, 0, 0, 1])

#intersection is 2, union is 3 #NOTE that these are binary values, so intersect would be where both are 1 (true)
#similarly union is where its either 1 and 1, or 1 and 0 or 0 and 1.

jac_sim = jaccard_score(A, B)
jac_dist = 1 - jac_sim


print(f"Jaccard Similarity: {jac_sim:.3f}")
print(f"Jaccard Distance: {jac_dist:.3f}")
```

```python
#cosine similarity
X = x.reshape(1, -1)
Y = y.reshape(1, -1) #reusing previous x and y values. note that [1,2,3] and [4,6,8] are basically the same direction so cos sim would approach 1
cos_sim = cosine_similarity(X, Y)[0, 0]
cos_dist = 1 - cos_sim
print(f"Cosine Similarity: {cos_sim:.3f}")
print(f"Cosine Distance: {cos_dist:.3f}")
```
