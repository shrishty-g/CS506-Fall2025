```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

#showing the pearsons coeff for y=x
x= np.arange(1,101)
y=x

# Compute Pearson correlation
r, p_value = pearsonr(x, y)
print(f"r = {r:.3f}, p-value = {p_value:.3f}")

# Plot
plt.scatter(x, y)
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
#showing that r is scaling invariant as well as shifting invariant
y= 3*x+5

r, p_value=pearsonr(x,y)
print(f"r={r: .3f}, pval={p_value: .3f}")
plt.scatter(x,y)
```

```python
#now for negative linear correlation
y=-2*x+10
r, p_value=pearsonr(x,y)
print(f"r={r: .3f}, pval={p_value: .3f}")
plt.scatter(x,y)
```

```python
#r value isnt always 1 or -1, lets see a weaker correlation
y= 2*x +np.random.normal(0,20,100)

r, p_value=pearsonr(x,y)
print(f"r={r: .3f}, pval={p_value: .3f}")
plt.scatter(x,y)
```

```python
#for no linear correlation, r=0

x= np.arange(-50,51)
y= x**2

r, p_value=pearsonr(x,y)
print(f"r={r: .3f}, pval={p_value: .3f}")
plt.scatter(x,y)
```
